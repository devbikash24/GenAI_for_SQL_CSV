```python
import chromadb
from pprint import pprint

# Initialize the ChromaDB PersistentClient
client = chromadb.PersistentClient(path="D:\Generative AI\generative ai project\QA_SQL_Tabular_data\data\chroma")

# List all collections in the database
collections = client.list_collections()
print("Collections in the database:")
for collection in collections:
    print(f"- {collection.name}")

# Choose a specific collection
collection_name = "titanic_data"
collection = client.get_collection(name=collection_name)

# Inspect the contents of the collection
# Retrieve all data in the collection (for small collections)
data = collection.get()
pprint(data)  # Pretty print the data

# Example: Get a subset of the data for large collections
# Example: Print out the retrieved data
print("Documents in the collection:")
for doc in data['documents'][0]:
    print(doc)
```

```python
# Example: Inspect metadata
metadata = data['metadatas']
pprint(metadata)

# Visualizing metadata distribution (example: count of different labels)
import pandas as pd

# Assuming metadata contains labels
labels = [meta['label'] for meta in metadata if 'label' in meta]
df_labels = pd.DataFrame(labels, columns=["label"])


```

```python
from sentence_transformers import SentenceTransformer
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
query_text = "what is four?"
query_embedding = embedding_model.encode(query_text).tolist()
results = collection.query(
    query_embeddings=[query_embedding],
    n_results=5  # Adjust the number of results you need
)
# Print the query results
# Display the results
for i, result in enumerate(results['documents']):
    print(f"Result {i + 1}:")
    print("Document:", result)
    print("Metadata:", results['metadatas'][i])
    print("Distance:", results['distances'][i])
    print("="*50)
```
